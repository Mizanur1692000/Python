# Box-Plot

# ViolinPlot

```python
sns.violinplot(x="SibSp", y="Pclass", data=df)
plt.show
```

```python
sns.violinplot(x="SibSp", y="Age", data=df)
plt.show
```

# Scatter-Plot

```python
df.plot(kind="scatter", x="SibSp", y="Age")
plt.title("Scatter-Plot")
plt.show
```

```python
df.plot(kind="scatter", x="Age", y="SibSp")
plt.show
```

```python
sns.set_style("whitegrid");
sns.pairplot(df, hue="Age",height=6)

plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="Survived",height=7)\
    .map(plt.scatter,"Pclass","Age")\
    .add_legend();
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="Survived",height=7)\
    .map(sns.distplot,"Age")\
    .add_legend();
plt.title("Histogram of ages of the travellers",fontsize=20)
plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="Survived",height=7)\
    .map(sns.distplot,"SibSp")\
    .add_legend();
plt.title("Histogram of ages of the travellers",fontsize=20)
plt.show();
```

# Pie-Chart

```python
df["Survived"].value_counts().plot(kind="pie", autopct="%1.2f%%", shadow=True, startangle=15, explode=(0.2,0))
plt.title("Pie Chart")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="pie", autopct="%1.2f%%", shadow=True, startangle=120, explode=(0.2,0))
plt.title("Pie Chart")
plt.show()
```
