# Lineplot

```python
import matplotlib.pyplot as plt
x=list(range(1,100,10))
y=list(range(1,50,5))
plt.plot(x,y,color="Green")
plt.xlabel("Range:1-100")
plt.ylabel("Range:1-50")
plt.title("Line Plot")
plt.show
```

```python
import matplotlib.pyplot as plt
x=[2,4,7,8,3,4,5,8]
y=[4,3,6,5,9,9,1,2]
plt.plot(x,y,color="Red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,4,7,8,3,4]
y=[4,3,6,5,9,9]
z=[4,2,6,3,8,4]
plt.plot(x,y,color="Cyan")
plt.plot(x,z,color="Blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Cyan","Blue"])
plt.title("Line Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,4,7,8,3,4]
y=[4,3,6,5,9,9]
z=[4,2,6,3,8,4]

plt.subplot(3,3,1)

plt.plot(x,y,color="Red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Red"])
plt.title("Line Plot")

plt.subplot(3,3,2)

plt.plot(x,z,color="Green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Red"])
plt.title("Line Plot")

plt.subplot(3,3,3)

plt.plot(y,z,color="Blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Blue"])
plt.title("Line Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,4,7,8]
y=[1,2,3,5]
z=[2,3,5,6]

plt.subplot(1,3,1)

plt.plot(x,y,color="Red")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Red"])
plt.title("Line Plot")

plt.subplot(1,3,2)

plt.plot(x,z,color="Green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Green"])
plt.title("Line Plot")

plt.subplot(1,3,3)

plt.plot(y,z,color="Blue")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Blue"])
plt.title("Line Plot")

plt.show()
```

```python
import matplotlib.pyplot as plt
x=list(range(1,100,10))
y=list(range(1,50,5))
plt.scatter(x,y,color="Orange",linewidth=2.0,linestyle=":",marker="o",alpha=.8)
plt.xlabel("Range:1-100")
plt.ylabel("Range:1-50")
plt.title("Line Plot")
plt.grid()
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
x=list(range(1,100,10))
y=list(range(1,50,5))
plt.scatter(x,y,color="Purple")
plt.xlabel("Range:1-100")
plt.ylabel("Range:1-50")
plt.title("Line Plot")
plt.show
```

```python
import matplotlib.pyplot as plt
x=[2,4,7,8,3,4]
y=[4,3,6,5,9,9]
plt.scatter(x,y,color="Blue",linewidth=2.0,linestyle=":",marker="o",alpha=.8)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,4,7,5,3,4]
y=[4,3,6,5,9,9]
z=[4,2,6,3,8,4]
plt.scatter(x,y,color="Red")
plt.scatter(x,z,color="Green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Red","Green"])
plt.title("Line Plot")
plt.grid()
plt.show()
```

# Bar Chart

```python
import matplotlib.pyplot as plt
x=["Mango","Orange","Apple","Lemon"]
y=[30,23,32,17]
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.title("Fruit Shop")
plt.bar(x,y,color="Green")

plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Mango","Orange","Apple","Lemon"]
y=[3,20,12,7]
plt.ylabel("Fruits")
plt.xlabel("Quantity")
plt.title("Fruit Shop")
plt.barh(x,y,color="Purple")

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x=["Lion","Dog","Bear","Wolf"]
y=[40,3,8,12]
plt.pie(y, labels=x, autopct="%1.2f%%", shadow=True, startangle=90)
plt.title("Ratio of the animals of a Jungle")
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

x=["Cat","Tiger","Bear","Deer"]
y=[4,3,8,34]
plt.pie(y, labels=x, autopct="%1.2f%%", shadow=True, startangle=90, explode=(0.2,0,0.5,0))
plt.title("Ratio of the animals of a Jungle")
plt.show()
```
